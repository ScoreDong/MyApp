```python
import pandas as pd 
import numpy as np 
import sys

file_name="\测试数据.xlsx"
file_path=r"C:\Users\YAO\1"
file=file_path+file_name
df = pd.read_excel(file,
                   usecols = "D,M,R:AK",
                   sheet_name= '订单明细')
df.head(10)
```

```python
# 设定保底和费率
设定保底=4.5
设定费率=15

# 测算保底和费率
测算保底=0 # 这个没保底的
测算费率=20

print(
"""
配送方式	费率	保底
美团快送	15.0%	4.5
美团专送	15.0%	4.5
到店取餐	5.0%	0.5
津贴联盟		-
代理商配送	16.0%	-按原价
商家自配	16.0%	-
""")
```

```python
# 版本0.21.0引入了infer_objects()方法，用于将具有对象数据类型的DataFrame的列转换为更具体的类型。

df['美团活动补贴']=df['美团活动补贴'].replace('-',0).astype(float)
df['用户支付配送费']=df['用户支付配送费'].replace('-',0).astype(float)
df['用户线上支付金额']=df['用户线上支付金额'].replace('-',0).astype(float)
df['费率']=df['费率'].str.strip("%").astype(float) # Python如何将百分号的字符转成数字
df['保底']=df['保底'].replace('-',0).astype(float) # Python强行转换格式
df['商家代金券美团补贴']=df['商家代金券美团补贴'].replace('-',0).infer_objects()
df['商家代金券商家补贴']=df['商家代金券商家补贴'].replace('-',0).infer_objects()
df['商家代金券用户支付']=df['商家代金券用户支付'].replace('-',0).infer_objects()
df['商家代金券抽佣比例']=df['商家代金券抽佣比例'].replace('-',0).infer_objects()

df.dtypes
```

```python
print("当前最大费率（代理商费率16，美团专送15）：",df['费率'].max(),"当前最大保底：",df['保底'].max())
print("""
费率	计数
""",
      df['费率'].value_counts())
```

```python
# 分类处理
df到店取餐=df[df['费率'].isin([5.0])] # 选取['费率']列包含数字0.5的行
df其他1=df[df['费率'].isin([0])] # 选取['费率']列包含数字0的行
df其他2=df[df['用户线上支付金额'].isin([0])] # 选取['费率']列包含数字0的行
df美团外卖=df[~df['费率'].isin([0,5.0,np.NaN])] # 通过~取反，选取不包含数字0和0.5的行
df美团专送=df[df['费率'].isin([15.0])] 
df美团代理商=df[df['费率'].isin([16.0])] 

df美团待测算=df美团专送

print(file_name,"美团服务费（含美团专送和代理商）：",-df美团外卖['平台服务费'].sum())
print(file_name,"美团专送服务费：",-df美团专送['平台服务费'].sum())
print(file_name,"美团代理商服务费：",-df美团代理商['平台服务费'].sum())
```

```python
df美团待测算['无保底服务费']=(df美团待测算['用户线上支付金额'] - df美团待测算['用户支付配送费'] + df美团待测算['美团活动补贴'])*设定费率/100
df美团待测算['有保底服务费']=df美团待测算['无保底服务费']
df美团待测算.loc[df美团待测算['无保底服务费']<设定保底,'有保底服务费']=设定保底 # 算保底

# 下面这个没保底的
df美团待测算['测算服务费']=(df美团待测算['用户线上支付金额'] - df美团待测算['用户支付配送费'] + df美团待测算['美团活动补贴'])*测算费率/100

df美团待测算['校验']=df美团待测算['有保底服务费']+df美团待测算['平台服务费'] # ['平台服务费']是负的

# df美团专送.loc[df美团专送['校验']>=1] 这样会出来一堆代理商配送，这种代理商配送的，是原价*费率，我们一定要谈掉，太坑爹了
```
